```python
#Gather the songs
#Remove songs already present in the hot_songs dataset
```

```python
def exclude_overlapping_songs(hot_songs, not_so_hot_songs, Artist, Song_title):
    """
    Exclude songs from hot_songs_df that are present in not_hot_songs_df.

    :param hot_songs: DataFrame containing hot songs.
    :param not_hot_songs: DataFrame containing not hot songs.
    :param Artist: Column name for artist in both DataFrames.
    :param Song_title: Column name for song title in both DataFrames.
    :return: DataFrame of hot songs excluding those present in not hot songs.
    """

    hot_songs['combined_key'] = hot_songs[Artist] + ' - ' + hot_songs[Song_title]
    not_so_hot_songs['combined_key'] = not_so_hot_songs[Artist] + ' - ' + not_so_hot_songs[Song_title]

    mask = hot_songs['combined_key'].isin(not_so_hot_songs['combined_key'])

    exclusive_hot_songs= hot_songs[~mask].copy()

    exclusive_hot_songs.drop(columns=['combined_key'], inplace=True)

    return exclusive_hot_songs
```

```python
exclusive_hot_songs = exclude_overlapping_songs(hot_songs, not_so_hot_songs, 'Artist', 'Song_title')

print("Number of exclusive hot songs:", len(exclusive_hot_songs))
```
